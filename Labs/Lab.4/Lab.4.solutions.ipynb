1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class Counter:
    def __init__(self, max_value):
        self.value = 0
        self.max_value = max_value

    def increment(self):
        if self.value < self.max_value:
            self.value += 1
        else:
            print("ERROR: Can't increment beyond this value")

    def reset(self):
        self.value = 0

    def get_value(self):
        return self.value
```

```python
counter = Counter(9)
for value in range(9):
    counter.increment()
    print(counter.get_value())

counter.reset()
print("Reset:", counter.get_value())
```

```python
class Rectangle:
    def __init__(self,length,width,x,y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y

    def get_length(self):
        return self.__length

    def set_length(self):
        self.__length = length

    def get_width(self):
        return self.__width

    def set_width(self):
        self.__width = width

    def get_x(self):
        return self.__x

    def set_x(self):
        self.__x = x

    def get_y(self):
        return self.__y

    def set_y(self):
        self.__y = y
    
    def area(self):
        area_ = (self.__length)*(self.__width)
        return area_
        
    def perimeter(self):
        perimeter_ = (2*(self.__length))+(2*(self.__width))
        return perimeter_
        
```

```python
class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.data = [[' '] * width for i in range(height)]

    def set_pixel(self, row, col, char='*'):
        self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]
    
    def clear_canvas(self):
        self.data = [[' '] * self.width for i in range(self.height)]
    
    def v_line(self, x, y, w, **kargs):
        for i in range(x,x+w):
            self.set_pixel(i,y, **kargs)

    def h_line(self, x, y, h, **kargs):
        for i in range(y,y+h):
            self.set_pixel(x,i, **kargs)
            
    def line(self, x1, y1, x2, y2, **kargs):
        slope = (y2-y1) / (x2-x1)
        for y in range(y1,y2):
            x= int(slope * y)
            self.set_pixel(x,y, **kargs)
            
    def display(self):
        print("\n".join(["".join(row) for row in self.data]))

class Shape:
    def __init__(self,x,y):
        self.__x = x
        self.__y = y

    def get_x(self):
        return self.__x

    def get_y(self):
        return self.__y

    def set_x(self, x):
        self.__x = x

    def set_y(self, y):
        self.__y = y

    def area(self):
        raise NotImplementedError
        
    def perimeter(self):
        raise NotImplementedError

    def points(self,n=16):
        raise NotImplementedError
        
    def inside_o(self):
        raise NotImplementedError

    def overlap_(self,shape_2):
        raise NotImplementedError

    def paint(self, canvas): pass
        
class Circle(Shape):
    def __init__(self,x,y,radius):
        Shape.__init__(self,x,y)
        self.__radius = radius

    def area(self):
        area = (math.pi)*((self.__radius)**2)
        return area

    def perimeter(self):
        return 2*(math.pi)*(self.__radius)

    def points(self, n=16):
         points = []
         for i in range(n):
             angle = (2 * math.pi * i) / n
             x = self.get_x() + math.cos(angle) * self.__radius
             y = self.get_y() + math.sin(angle) * self.__radius
             points.append((x, y))
         return points

    def inside_o(self,x,y):
        distance = math.sqrt((x-self.get_x())**2+(y-self.get_y())**2)
        if distance <= self.__radius:
            return print("this point is within the circle")
        else:
            return print("this point is not within the circle")

    def overlap_(self,shape_2):
        distance_c = math.sqrt((self.get_x()-shape_2.get_x())**2+(self.get_y()-shape_2.get_y())**2)
        return distance_c < (self.__radius + shape_2.__radius)

    def paint(self, canvas, char='*'):
        center_x = self.get_x()
        center_y = self.get_y()
        radius_squared = self.__radius**2
        for i in range(int(center_x - self.__radius), int(center_x + self.__radius)+1):
            for j in range(int(center_y - self.__radius), int(center_y + self.__radius)+1):
                if (i - center_x) ** 2 + (j - center_y) ** 2 <= radius_squared:
                    canvas.set_pixel(j, i, char)
        

class Rectangle(Shape):
    def __init__(self,x,y,length,width):
        Shape.__init__(self,x,y)
        self.__length = length
        self.__width = width

    def area(self):
        area_ = (self.__length)*(self.__width)
        return area_
        
    def perimeter(self):
        perimeter_ = (2*(self.__length))+(2*(self.__width))
        return perimeter_

    def points(self, n=16):
        points = []
        for i in range(n):
            if i < self.__length:  # Top side
                points.append((self.x + i, self.y))
            elif i < self.__length + self.__width:  # Right side
                points.append((self.x + self.__length, self.y + i - self.__length))
            elif i < 2 * self.__length + self.__width:  # Bottom side
                points.append((self.x + self.__length - (i - self.__length - self.__width), self.y + self.__width))
            else:  # Left side
                points.append((self.x, self.y + 2 * self.__length + self.__width - i))
        return points

    def inside_o(self,x,y):
        half_length = self.__length / 2
        half_width = self.__width / 2
        return abs(x - self.get_x()) <= half_length and abs(y - self.get_y()) <= half_width
        
    def overlap_(self,shape_2):
        return (abs(self.__x()-shape_2.__x())*2 < (self.__length + shape_2.__length) and abs(self.__y() - shape_2.__y()) * 2 < (self.__width + shape_2.__width))
    
    def paint(self, canvas, char='*'):
        x = self.get_x()
        y = self.get_y()
        
        for i in range(int(x - self.__length/2), int(x + self.__length/2)):
            for j in range(int(y - self.__width/2), int(y + self.__width/2)):
                canvas.set_pixel(i, j, char)

class Triangle(Shape):
    def __init__(self,side_1,base,height,x,y):
        Shape.__init__(self,x,y)
        self.__side_1 = side_1
        self.__base = base
        self.__height = height

    def area(self):
        return (1/2)*(self.__base)*(self.__height) 

    def perimeter(self):
        return (self.__side_1)+(self.__height)+(self.__base)


    def points(self, n=16):
        points = []

        # Adding points along the base
        for i in range(n):
            x = self.get_x() + i * self.__base / n
            y = self.get_y()
            points.append((x, y))

        # Adding points along the height (right side)
        for i in range(n):
            x = self.get_x()
            y = self.get_y() + i * self.__height / n
            points.append((x, y))

        # Adding points along the hypotenuse
        for i in range(n):
            x = self.get_x() + i * self.__base / n
            y = self.get_y() + (self.__height - i * self.__height / self.__base)
            points.append((x, y))

        return points

    def inside_o(self,x,y):
        return 0 <= x <= self.__base and 0 <= y <= self.__height

    def overlap_(self, shape_2):
        if isinstance(shape_2, Triangle):
            vertices = self.points()  
            for vertex in vertices:
                x, y = vertex
                if shape_2.inside_o(x, y):
                    return True
            
            vertices_shape_2 = shape_2.points()  
            for vertex in vertices_shape_2:
                x, y = vertex
                if self.inside_o(x, y):
                    return True
            
            return False
        else:
            return False

    def paint(self, canvas, char='*'):
        points = self.points()
        for point in points:
            x, y = point
            canvas.set_pixel(int(x), int(y), char)

class CompoundShape(Shape):
    def __init__(self, shapes):
        self.shapes = shapes
        x = min([shape.get_x() for shape in shapes])
        y = min([shape.get_y() for shape in shapes])
        Shape.__init__(self, x, y)

    def area(self):
        return sum([shape.area() for shape in self.shapes])

    def perimeter(self):
        return sum([shape.perimeter() for shape in self.shapes])

    def points(self):
        all_points = [shape.points() for shape in self.shapes]
        return [point for sublist in all_points for point in sublist]

    def inside_o(self, x, y):
        return any([shape.inside_o(x, y) for shape in self.shapes])

    def overlap_(self, shape_2):
        return any([shape.overlap_(shape_2) for shape in self.shapes])

    def paint(self, canvas, char='*'):
        for shape in self.shapes:
            shape.paint(canvas, char)
```

```python
class RasterDrawing:
    def __init__(self):
        self.shapes=dict()
        self.shape_names=list()
        
    def add_shape(self,shape):
        if shape.name == "":
            shape.name = self.assign_name()
        
        self.shapes[shape.name]=shape
        self.shape_names.append(shape.name)

    def update(self, canvas):
        canvas.clear_canvas()
        self.paint(canvas)
        
    def paint(self,canvas):
        for shape_name in self.shape_names:
            self.shapes[shape_name].paint(canvas)
            
    def assign_name(self):
        name_base="shape"
        name = name_base+"_0"
        
        i=1
        while name in self.shapes:
            name = name_base+"_"+str(i)
            
        return name
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __repr__(self):
        return f"Point({self.x}, {self.y})"


class Line:
    def __init__(self, start, end):
        self.start = start
        self.end = end

    def __repr__(self):
        return f"Line({self.start}, {self.end})"


class RasterDrawing:
    def __init__(self):
        self.shapes = []

    def add_shape(self, shape):
        self.shapes.append(shape)

    def update(self, canvas):
        canvas.clear_canvas()
        self.paint(canvas)

    def paint(self, canvas):
        for shape in self.shapes:
            shape.paint(canvas)

    def save(self, filename):
        with open(filename, 'w') as f:
            for shape in self.shapes:
                f.write(repr(shape) + '\n')

    def load(filename):
        drawing = RasterDrawing()
        with open(filename, 'r') as f:
            for line in f:
                shape = eval(line.strip())
                drawing.add_shape(shape)
        return drawing
```
